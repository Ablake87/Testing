```python
import lvm_read
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filename1 = 'Lab_5_10.lvm'
lvm = lvm_read.read(filename1, read_from_pickle=False)

plt.grid()

plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")

plt.plot(lvm[1] ['data'], label='DC');
plt.plot(lvm[2] ['data'], label='AC');

plt.legend(loc="best")
```
